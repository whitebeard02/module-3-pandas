```python
import pandas as pd
import numpy as np

np.random.seed(10)

departments = ['HR', 'Engineering', 'Sales', 'Finance', 'Marketing']
n = 200

data = {
    'Employee_ID': range(1, n + 1),
    'Department': np.random.choice(departments, n),
    'Age': np.random.randint(22, 55, n),
    'Salary': np.random.randint(30000, 120000, n),
    'Experience': np.random.randint(1, 20, n),
    'Performance': np.random.randint(40, 100, n),
    'Attrition': np.random.choice(['Yes', 'No'], n, p=[0.3, 0.7])
}

df = pd.DataFrame(data)
df.to_csv('employees.csv', index=False)
print("Dataset created successfully")
```

```python
import pandas as pd

# loading the dataset
df = pd.read_csv('employees.csv')

# Task 2 - first 10 records

print("First 10 Records:")
print(df.head(10))

# Task 3 - averages

print("\nAverage Salary   :", round(df['Salary'].mean(), 2))
print("Average Age      :", round(df['Age'].mean(), 2))
print("Average Experience:", round(df['Experience'].mean(), 2))


# Task 4 - employee count per department

print("\nEmployee Count by Department:")
print(df['Department'].value_counts())


# Task 5 - department with highest attrition

attrition_df = df[df['Attrition'] == 'Yes']
highest_attrition = attrition_df['Department'].value_counts().idxmax()
print("\nDepartment with Highest Attrition:", highest_attrition)


# Task 6 - filtering employees

high_salary = df[df['Salary'] > 50000]
print("\nEmployees with Salary > 50000:", len(high_salary))

experienced = df[df['Experience'] > 5]
print("Employees with Experience > 5 years:", len(experienced))


# Task 7 - sort by salary descending

sorted_df = df.sort_values('Salary', ascending=False)
print("\nTop 10 Highest Paid Employees:")
print(sorted_df[['Employee_ID', 'Department', 'Salary']].head(10))


# Task 8 - group by department

print("\nAverage Salary and Performance by Department:")
grouped = df.groupby('Department')[['Salary', 'Performance']].mean().round(2)
print(grouped)


# Task 9 - Risk Level column

def assign_risk(row):
    if row['Attrition'] == 'Yes':
        return 'High Risk'
    elif row['Performance'] < 60:
        return 'Medium Risk'
    else:
        return 'Low Risk'

df['Risk_Level'] = df.apply(assign_risk, axis=1)

print("\nRisk Level Distribution:")
print(df['Risk_Level'].value_counts())


# Export cleaned dataset

df.to_csv('employees_cleaned.csv', index=False)
print("\nCleaned dataset exported successfully")
```
